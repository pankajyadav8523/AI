# Reinforcement Learning
Implementing reinforcement learning for playig tic-tac-toe game.

```python
# Importing required modules.
import random
import json
from utility import *           #Importing my own created module in main directory.
```

```python
# Creating tree initialized with stat_state
tree = {'NNNNNNNNN': 0}
alpha = 0.365248        # Choosing alpha.
```

```python
def learning(state, player, value = 0):
    ''''''
    for i in range(9):
        if state[i] == 'N':
            next_state = create_state(state, player, i + 1)
            status = is_terminal(next_state)
            if next_state not in tree:
                tree[next_state] = 0
            else:
                tree[next_state] = value
            if status[0]:
                tree[next_state] = status[1]
                tree[state] = tree[state] + alpha * (tree[next_state] - tree[state])
                return
            value = tree[state] + alpha * (tree[next_state] - tree[state])
            tree[state] = value
            learning(next_state, 'O' if player == 'X' else 'X', value)
learning('NNNNNNNNN', 'O')
print(tree)
len(tree)
```

```python
# Iteratively learning the value for states.
for i in range(1000):
    lst = list(tree.keys())
    r = random.randint(0, len(lst) - 1)
    k = lst[r]
    learning(k, 'O', tree[k])
```

```python
# Writing learned parameter file to the Minmax json file.
file_path = "Reinforcement.json"
with open(file_path, 'w') as json_file:
    json.dump(tree, json_file)
```
