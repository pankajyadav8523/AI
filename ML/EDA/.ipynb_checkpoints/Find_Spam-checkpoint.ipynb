```python
import random as r
```

```python
import pandas as pd
```

```python
df = pd.read_csv("SMSSpamCollection .csv")
```

```python
df
```

```python
df1 = df[df["EmailType"] == "ham"]
```

```python
df1
```

```python
df1["Text"]
```

```python
Total_ham_text_list = []
for index,row in df1.iterrows():
    ham = row["Text"]
    Total_ham_text_list.append(ham)
```

```python
df2 = df[df["EmailType"] == "spam"]
```

```python
df2
```

```python
Total_spam_text_list = []
for index,row in df2.iterrows():
    spam = row["Text"]
    Total_spam_text_list.append(spam)
```

```python
r.shuffle(Total_ham_text_list)
print(Total_spam_text_list)
print(len(Total_spam_text_list))
```

```python
print(Total_ham_text_list)
```

```python
random_ham_text = r.sample(Total_ham_text_list, 747)
r.shuffle(random_ham_text)
```

```python
# For testing we have to choose 20% of ham and spam
twenty_percent_ham = Total_ham_text_list[598:]
twenty_percent_spam = Total_spam_text_list[598:]
```
