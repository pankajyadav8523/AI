```python
with open('0.edges', 'r') as file:
    lines = file.readlines() # convert file into list of lines
    for line in lines:       # iterate over each edge
        edges = line.split() # convert line into list of nodes that's called edge.
        d[edges[0]] = edges[1]
        nodes.append(edges[0])
        nodes.append(edges[1])
        lst_edges.append(edges)
        
```

```python
# Create the adjacency matrix

adj_matrix = []
for row in d:
    matrix_element = []
    for col in d:
        edge = [row, col]
        if edge in lst_edges and row == col:
            matrix_element.append(0)
        elif edge in lst_edges and row != col:
            count += 1
            matrix_element.append(1)
        else:
            matrix_element.append(0)
        
        
    adj_matrix.append(matrix_element)
    
    
print(adj_matrix)
```
